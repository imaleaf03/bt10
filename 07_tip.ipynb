### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv')
tips
```

### Step 4. Delete the Unnamed 0 column

```python
tips.drop('Unnamed: 0', axis=1, inplace=True)
```

### Step 5. Plot the total_bill column histogram

```python
plt.hist(tips['total_bill'])
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Histogram of Total Bill')
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.scatter(tips['total_bill'], tips['tip'])
plt.title('Relationship between Total Bill and Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.grid(True)
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
colors = tips['size']
plt.figure(figsize=(10, 8))
scatter = plt.scatter(tips['total_bill'], tips['tip'], c=colors, cmap='viridis', alpha=0.5)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Relationship between Total Bill, Tip, and Size')
plt.colorbar(scatter, label='Size')
plt.grid(True)
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
avg_total_bill = tips.groupby('day')['total_bill'].mean().reset_index()
plt.bar(avg_total_bill['day'], avg_total_bill['total_bill'])
plt.xlabel('Day')
plt.ylabel('Average Total Bill')
plt.title('Average Total Bill by Day')
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
plt.scatter(tips[tips['sex'] == 'Female']['tip'], tips[tips['sex'] == 'Female']['day'],
            color='blue', label='Female', alpha=0.5)

plt.scatter(tips[tips['sex'] == 'Male']['tip'], tips[tips['sex'] == 'Male']['day'],
            color='red', label='Male', alpha=0.5)

plt.xlabel('Tip')
plt.ylabel('Day')
plt.title('Scatter Plot of Tip vs Day by Sex')
plt.legend()
plt.grid(True)
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(x='day', y='total_bill', hue='time', data=tips, palette='Set2')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Total Bill per Day, Differentiated by Time')
plt.legend(title='Time')
plt.grid(True)
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
dinner_tips = tips[tips['time'] == 'Dinner']['tip']
lunch_tips = tips[tips['time'] == 'Lunch']['tip']
plt.subplot(1, 2, 1)
plt.hist(dinner_tips)
plt.xlabel('Tip')
plt.ylabel('Frequency')
plt.title('Tip Distribution for Dinner')

plt.subplot(1, 2, 2)
plt.hist(lunch_tips)
plt.xlabel('Tip')
plt.ylabel('Frequency')
plt.title('Tip Distribution for Lunch')

plt.tight_layout()
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
male = tips[tips['sex'] == 'Male']
female = tips[tips['sex'] == 'Female']

plt.subplot(1, 2, 1)
plt.scatter(male[male['smoker'] == 'Yes']['total_bill'], male[male['smoker'] == 'Yes']['tip'], label='Smoker')
plt.scatter(male[male['smoker'] == 'No']['total_bill'], male[male['smoker'] == 'No']['tip'], label='Non-Smoker')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Male: Total Bill vs Tip')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(female[female['smoker'] == 'Yes']['total_bill'], female[female['smoker'] == 'Yes']['tip'], label='Smoker')
plt.scatter(female[female['smoker'] == 'No']['total_bill'], female[female['smoker'] == 'No']['tip'], label='Non-Smoker')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Female: Total Bill vs Tip')
plt.legend()
plt.show()
```
