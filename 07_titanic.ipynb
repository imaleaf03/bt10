# Visualizing the Titanic Disaster

### Introduction:

This exercise is based on the titanic Disaster dataset avaiable at [Kaggle](https://www.kaggle.com/c/titanic).  
To know more about the variables check [here](https://www.kaggle.com/c/titanic/data)


### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv)

### Step 3. Assign it to a variable titanic

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv')
titanic
```

### Step 4. Set PassengerId as the index

```python
titanic.set_index('PassengerId')
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
gender = titanic['Sex'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(gender, labels=gender.index, autopct='%1.1f%%', startangle=140)
plt.title('Male/Female Proportion')
plt.axis('equal')
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
plt.scatter(titanic[titanic['Sex'] == 'male']['Age'], titanic[titanic['Sex'] == 'male']['Fare'], color='blue', label='Male')
plt.scatter(titanic[titanic['Sex'] == 'female']['Age'], titanic[titanic['Sex'] == 'female']['Fare'], color='red', label='Female')
plt.title('Scatterplot of Fare Paid vs Age')
plt.xlabel('Age')
plt.ylabel('Fare Paid')
plt.legend()
plt.show()
```

### Step 7. How many people survived?

```python
titanic['Survived'].sum()
```

### Step 8. Create a histogram with the Fare payed

```python
plt.hist(titanic['Fare'])
plt.title('Histogram of Fare Paid')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
